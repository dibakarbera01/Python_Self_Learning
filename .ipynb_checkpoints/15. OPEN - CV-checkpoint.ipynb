```python
import cv2
```

```python

cap = cv2.VideoCapture(0)

while True:
    ret,frame = cap.read()

    if ret == False:
        continue

    cv2.imshow("Video Frame",frame)

    key_pressed = cv2.waitkey(1) & 0xFF
    if key_pressed == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
