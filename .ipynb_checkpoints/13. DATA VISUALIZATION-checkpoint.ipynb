## Bar Graphs

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python


plt.bar([0,1,2],[10,20,15]) #current year
plt.bar([0,1,2],[20,10,12]) # next year
plt.show()
```

```python

x_coordinates = np.array([0,1,2])*2

plt.bar(x_coordinates-0.25,[10,20,15],width=0.5,label="current year",tick_label=["Gold","Platinum","Silver"]) #current year
plt.bar(x_coordinates+0.25,[20,10,12],width=0.5,label="next year")      # next year
plt.title("Metal price Comparison")
plt.xlabel("Metal")
plt.ylabel("Price")
plt.ylim(0,40)
plt.xlim(-2,5)
plt.legend()
plt.show()
```

```python
plt.style.use("dark_background")
x_coordinates = np.array([0,1,2])*2

plt.bar(x_coordinates-0.25,[10,20,15],width=0.5,label="current year",tick_label=["Gold","Platinum","Silver"],color="green")
plt.bar(x_coordinates+0.25,[20,10,12],width=0.5,label="next year",color="orange")      
plt.title("Metal price Comparison")
plt.xlabel("Metal")
plt.ylabel("Price")
plt.legend()
plt.show()
```

## Data Visualisation Challenge


Given a Movies Dataset. Your tasks is to make a visualisation that-
- plots the length of movie titles names on x-axis
- Frequency counts on y-axis i.e number of movies having 'x' characters in their title

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# read a csv file
df = pd.read_csv("dataset/movie/movie_metadata.csv")
```

```python
df.head(n=10)
```

```python
df.columns
```

```python
titles = df.get('movie_title')
```

```python
print(titles)
```

```python
print(type(titles))
```

```python
titles = list(df.get('movie_title'))
```

```python
print(type(titles))
```

```python
print(titles[:5])
```

```python
print(titles[0][:-3])
```

```python
freq_titles = {}

for title in titles:
    length = len(title)
     
    if freq_titles.get(length) is None:
        freq_titles[length] = 1
    else:
        freq_titles[length] += 1
```

```python
print(freq_titles)
```

```python
X = np.array(list(freq_titles.keys()))
Y = np.array(list(freq_titles.values()))
# print(X,Y)
```

```python
plt.scatter(X,Y)
plt.xlabel("Length of Movie Title")
plt.ylabel("Freq Count")
plt.title("Movie Data Visualisation ")
plt.show()
```

### DATA VISUALIZATION - NORMAL DISTRIBUTION II

#### Histogram

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# generate a normnal distribution
u = 5
sigma = 1
# such a distribution is also called standard normal distribution

vals = u + sigma*np.random.randn(1000)
print(vals.shape)

plt.hist(vals,100)
plt.show()
```

```python
vals = np.round(vals)
z = np.unique(vals,return_counts=True)
print(z)
```

```python
x = vals
y = np.zeros(x.shape)

plt.scatter(x,y)
plt.show()
```

## Multivariate Normal Distribution

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# bivariate
mean = np.array([0.0,0.0])
cov = np.array([[1,0.8],[0.8,1]])

mean2 = np.array([5.0,6.0])
cov2 = np.array([[1.3,0.2],[0.2,1.1]])

dist = np.random.multivariate_normal(mean,cov,500)
dist2 = np.random.multivariate_normal(mean2,cov2,500)
print(dist.shape)
plt.scatter(dist[:,0],dist[:,1])
plt.scatter(dist2[:,0],dist[:,1])
plt.show()
```

## normalisation // standradisation

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dfx = pd.read_csv('dataset/linear Regression/linearX.csv')
dfy =  pd.read_csv('dataset/linear Regression/linearY.csv')

x = dfx.values
y = dfy.values

#normalise
x2 = (x-x.mean())/x.std()

plt.scatter(x,y,label="Raw data")
plt.scatter(x2,y,color='orange',label="normalised data")
plt.legend()
plt.show()
```
