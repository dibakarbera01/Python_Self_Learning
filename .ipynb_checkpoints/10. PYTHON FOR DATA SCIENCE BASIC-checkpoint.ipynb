# Data visualisation

```python
import matplotlib.pyplot as plt
```

```python
x = np.array([1,2,3,4,5])
y = x**2 + 3
```

```python
plt.style.use('seaborn')
```

```python
print(x,y)
```

```python
plt.plot(x,x**2+3,label="stock price")
plt.plot(x,x**2,label="stock price")
plt.title('My line plot')
plt.xlabel("Timestap")
plt.ylabel("price")
plt.legend()
plt.show()
```

```python
plt.scatter(x,x**2+3,label="stock price")
plt.scatter(x,x**2,label="stock price")
plt.title('My scatter plot')
plt.xlabel("Timestap")
plt.ylabel("price")
plt.legend()
plt.show()
```

# Working with OpenCv & images

```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('image/bulbasaur.jpg')
rgb_image = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
print(img.shape)
```

```python
plt.imshow(rgb_image)
plt.axis("off")
plt.style.use('seaborn')
plt.imshow(img)
```

```python
crop_image = rgb_image[250:5000,150:4500,:]
plt.imshow(rgb_image)
plt.imshow(rgb_image)
plt.title("bulbasuar" + str(crop_image.shape))
plt.show()
```

```python
crop_image.shape
```

```python
total_pixels = crop_image.shape[0]*crop_image.shape[1]
```

```python
print(total_pixels)
```
