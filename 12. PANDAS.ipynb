```python
for t in titles:
    print(t)
```

```python
for t in titles:
    length = len(t)
    print(length)
```

```python
freq_title = {}

for t in titles:
    length = len(t)
  
    
    try:
        freq_title[length] += 1
    except:
        freq_title[length] = 1
        
print(freq_title)    
        
```

```python
X = np.array(list(freq_title.keys()))
Y = np.array(list(freq_title.values()))
```

```python
plt.scatter(X,Y)
plt.xlabel("Movie title len")
plt.ylabel("no. of Movies")
plt.show()
```
